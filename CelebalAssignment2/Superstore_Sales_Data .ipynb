## STEP 1: Import Libraries

```python
import sqlite3
import pandas as pd
import numpy as np
from datetime import datetime

print("Libraries imported")
```

## STEP 2: Load CSV file

```python
data = pd.read_csv("Sample - Superstore.csv", encoding="latin1")
```

## Create SQLite database

```python
conn = sqlite3.connect("superstore.db")
cursor=conn.cursor()
```

## Loading Data into Table

```python
# Creating table superstore
data.to_sql("superstore", conn, if_exists="replace", index=False)
print("Dataset Loaded Successfully")
```

```python
print(""" BRIEF INSIGHTS!!
1. The Superstore dataset was successfully loaded into SQLite.
2. Storing data in a SQL database enables efficient querying, filtering, and analysis.""")
```

```python
# query execution function
def run_query(query):
    return pd.read_sql(query,conn)
```

## STEP3. WHERE Filters

```python
print("Filtering the rows by region as east")

query = """SELECT *
FROM superstore
WHERE Region = 'East'
"""
pd.read_sql(query, conn)
```

```python
print(""" BRIEF INSIGHTS!!
1. TThis filter isolates sales transactions from the East region.
2. It helps analyze regional sales performance.""")
```

```python
print("Filtering the rows by Category = Furniture")

query ="""SELECT *
FROM superstore
WHERE Category = 'Furniture'"""
pd.read_sql(query, conn)
```

```python
print(""" BRIEF INSIGHTS!!
1. Furniture products can be analyzed separately from other categories.
2. Useful for understanding category-specific demand and profitability""")
```

```python
print("Filtering the rows by Sales > 1500")

query ="""SELECT *
FROM superstore
WHERE Sales > 1500"""
pd.read_sql(query, conn)
```

```python
print(""" Brief Insights!!! 
1. High-value transactions are identified.
2. These orders contribute significantly to overall revenue.""")
```

```python
print("Filtering the rows by Orders in 2015")

query= """SELECT *
FROM superstore
WHERE "Order Date" LIKE '%2015'"""
pd.read_sql(query, conn)
```

```python
print(""" Brief Insights!!! 
1.Filtering by year helps analyze historical sales trends.
2. Useful for year-over-year performance comparisons..""")
```

## STEP4: GROUP BY Aggregations

```python
print("Grouping by Total Sales by Region ")
query = """SELECT Region,
       ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY Region """

pd.read_sql(query, conn)
```

```python
print(""" Brief Insights!!! 
1. Reveals which region generates the highest revenue.
2. Helps management focus on high-performing markets. """)
```

```python
print("Grouping by Total Quantity by Category")
query = """SELECT Category,
       SUM(Quantity) AS Total_Quantity
FROM superstore
GROUP BY Category"""

pd.read_sql(query, conn)
```

```python
print("Grouping by Average Profit by Category ")
query = """SELECT Category,
       ROUND(AVG(Profit),2) AS Avg_Profit
FROM superstore
GROUP BY Category"""

pd.read_sql(query, conn)
```

## STEP5: Sort and Limit Results

```python
print(" Top 5 Products by Sales ")
query = """SELECT "Product Name",
       ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY "Product Name"
ORDER BY Total_Sales DESC
LIMIT 5 """

pd.read_sql(query, conn)
```

```python
print(""" Brief Insights!!! 
Highlights the highest revenue-generating products.
These products may require priority inventory management.""")
```

```python
print(" Top Categories by Profit ")
query = """SELECT Category,
       ROUND(SUM(Profit),2) AS Total_Profit
FROM superstore
GROUP BY Category
ORDER BY Total_Profit DESC"""

pd.read_sql(query, conn)
```

# STEP 6: Business Use Cases

```python
print("Monthly Sales Trend")
query = """SELECT SUBSTR("Order Date",-4) AS Year,
       SUBSTR("Order Date",1,2) AS Month,
       ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY Year, Month
ORDER BY Year, Month"""

pd.read_sql(query, conn)
```

```python
print(""" Brief Insights!!! 
Displays sales performance over time.
Helps identify seasonal demand patterns and growth trends.""")
```

```python
print("Top 5 Customers")
query = """SELECT "Customer Name",
       ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY "Customer Name"
ORDER BY Total_Sales DESC
LIMIT 5"""

pd.read_sql(query, conn)
```

```python
print("Find Duplicate Orders")

query = """SELECT "Order ID",
       COUNT(*) AS Duplicate_Count
FROM superstore
GROUP BY "Order ID"
HAVING COUNT(*) > 1 """

pd.read_sql(query, conn)
```

# STEP 7: Validate Results

```python
print(" Null Values ")
query = """SELECT
COUNT(*) -
COUNT("Order ID") AS Missing_OrderID,
COUNT(*) -
COUNT(Sales) AS Missing_Sales
FROM superstore """

pd.read_sql(query, conn)
```

```python
print("Row Count Check")
query = """SELECT COUNT(*) AS Total_Rows
FROM superstore """

pd.read_sql(query, conn)
```

```python
print("""Brief Insights!!!
1. Confirms that no records were lost during data loading or transformation.
2. Ensures consistency throughout the analysis process.""")
```

```python
print("Distinct Customers")
query = """SELECT COUNT(DISTINCT "Customer ID") AS Total_Customers
FROM superstore """

pd.read_sql(query, conn)
```

```python
print("""Brief Insights!!!
1. Shows the number of unique customers in the dataset.
2. Helps estimate customer base size and market reach.""")
```
